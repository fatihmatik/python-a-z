```python
#Importing the libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
display(df.head(3))
print(df.shape)
```

```python
countries_df = pd.DataFrame({"CountryCode":Codes_2012_Dataset,
                             "CountryName":Countries_2012_Dataset,
                             "Region":Regions_2012_Dataset})
countries_df.head(3)
```

```python
merged_df = pd.merge(left=merged_df, right=countries_df)
merged_df.head(3)
```

```python
l = [merged_df.columns[itr].replace(" ","") for itr in range(0,len(merged_df.columns),1)]
print(l)
```

```python
merged_df= merged_df.reindex(columns=['CountryName', 'CountryCode', 'Region', 'Birthrate', 'Internetusers', 'IncomeGroup', 'LifeEx1960', 'LifeEx2013'])
display(merged_df.head(3))
print(merged_df.shape)
```

```python
sns.lmplot(data=merged_df, x="LifeEx1960", y="Birthrate", hue="Region",
           fit_reg=False, height=4, aspect=2, scatter_kws={"alpha": 0.5})
plt.title("Life Expectancy of 1960 versus The Fertility Rate")
plt.grid(True)
plt.show()
```

```python
sns.lmplot(data=merged_df, x="LifeEx2013", y="Birthrate", hue="Region",
           fit_reg=False, height=4, aspect=2, scatter_kws={"alpha": 0.5})
plt.title("Life Expectancy of 2013 versus The Fertility Rate")
plt.grid(True)
plt.show()
```

```python
corrMatrix = merged_df.corr(numeric_only=True)
corrMatrix
plt.figure(figsize=(6,3))
sns.heatmap(data=corrMatrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between the Columns")
plt.show()
```

#### Interpretation
We can see that, aside from the African countries (mostly), life expectancy has improved and birth rates have decreased for countries in other continents. While the statistics remained the same for European countries. The graphs tell us that there's a correlation between the life expectancy of a country and the birth rates.

From the heatmap graph, we can see that there's a negative correlation between the Birthrate and the Life Expectancy features, as well as with the Internet Users feature.
