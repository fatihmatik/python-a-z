```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("P4-Section6-Homework-Dataset.csv", encoding="cp437")
df.columns = [column.replace(" ","") for column in df.columns]
```

```python
df.head(3)
```

```python
fig, axs = plt.subplots()

fig.set_figwidth(10)
fig.set_figheight(8)
sns.set_style("darkgrid")


filter1 = df["Genre"].isin(["action","comedy","adventure","animation","drama"])
filter2 = df["Studio"].isin(["Buena Vista Studios","Sony","Universal","WB","Paramount Pictures","Fox"])
df = df[filter1 & filter2]

sns.boxplot(data=df, x="Genre", y="Gross%US",
            color="#ccc",
            fliersize=0,
            ax=axs,
            )


custom_palette = {"Buena Vista Studios": "blue",
                  "Sony": "green",
                  "Universal": "red",
                  "WB": "purple",
                  "Paramount Pictures": "gold",
                  "Fox": "cyan"}

sns.stripplot(data=df,
                x="Genre",
                y="Gross%US",
                hue="Studio",
                jitter=0.2,
                ax=axs,
                size=6,
                alpha=0.6,
                palette=custom_palette,
                )

axs.set_title("Domestic Gross % by Genre", fontsize=28)
axs.set_xlabel("Genre", fontsize=24)
axs.set_ylabel("Gross % US", fontsize=24)
axs.legend(bbox_to_anchor=(1,1), frameon=False, fontsize=14)
axs.tick_params(axis='x', labelsize=16)
axs.tick_params(axis='y', labelsize=16)

plt.grid(True)
plt.show()
```
